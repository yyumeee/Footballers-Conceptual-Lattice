```python
import pandas as pd
#from arulespy.arules import Transactions, apriori, parameters
from sklearn.preprocessing import QuantileTransformer
import numpy as np

df = pd.read_csv('gk_factors.csv')
ind = ['player', 'age', 'nationality', 'position', 'team','top5', 'minutes_90s']
df.set_index(ind, inplace = True)
df
```

```python
ind = df.index
cols = df.columns
tr = QuantileTransformer(n_quantiles = 100)
df = tr.fit_transform(df)
df = pd.DataFrame(df, index = ind, columns = cols)
df
```

```python
cols = df.columns
for col in cols:
    df[f'{col} = Low'] = np.where(df[col] < 0.25, True, False)
    df[f'{col} = Average'] = np.where((df[col] >= 0.25) & (df[col] <= 0.75), True, False)
    df[f'{col} = High'] = np.where(df[col] > 0.75, True, False)
    df.drop(col, axis = 1, inplace = True)
df
```

```python
df.to_csv('gk_concepts.csv')
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

itemsets_freq = apriori(df, min_support = 0.05, use_colnames = True)
itemsets_freq.sort_values(by = 'support', ascending = False)
```

```python
arules = association_rules(itemsets_freq, metric="confidence", min_threshold=0.8)
arules.sort_values(by = 'lift', ascending = False).head()
```

```python
for rule in arules.iterrows():
    lhs = rule[1].antecedents
    rhs = rule[1].consequents

    lhs_cond = df[list(lhs)].all(axis=1)
    rhs_cond = df[list(rhs)].all(axis = 1) == False

    exceptions = df[lhs_cond & rhs_cond]
    exc = []
    for e in exceptions.iterrows():
        exc.append(e[0][0])

    print(f'rule {rule[1].name + 1}: {list(lhs)} -> {list(rhs)}')
    print(f'#players with lhs: {len(df[lhs_cond])}')
    print(f'#players respecting the rule: {rule[1].support * len(df)}')
    print(f'exceptions: {exc}')
    print('-'*40)
```
