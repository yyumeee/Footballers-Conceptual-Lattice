```python
import pandas as pd
import os

cart = 'scraped_data'

seriea = pd.read_csv(f'{cart}\\Serie-A.csv')
seriea
```

```python
rename = {'xg_net': 'G-xg', 'npxg_net': 'npG-npxg', 'assisted_shots': 'key_passes', 'xg_assist_net':'A-xa'}
```

```python
togliere = ['games_starts', 'minutes','birth_year','goals_assists', 'goals_pens', 'pens_made', 'goals_assists_per90', 'goals_pens_per90', 
            'goals_assists_pens_per90','xg_xg_assist_per90', 'npxg_xg_assist_per90', 'shots', 'shots_on_target', 'passes_completed_short',
            'passes_completed_medium', 'passes_completed_long', 'corner_kicks_straight', 'sca', 'gca', 'gca_per90',
            'gca_passes_live', 'gca_passes_dead', 'gca_take_ons', 'gca_shots', 'gca_fouled', 'gca_defense', 'challenges_lost',
            'blocked_shots', 'blocked_passes', 'tackles_interceptions', 'take_ons_won', 'take_ons_tackled', 'cards_yellow_red', 'matches']
```

```python
len(togliere)
```

```python
ind = ['player', 'nationality', 'position', 'team', 'age']
seriea = seriea.set_index(ind)
seriea.drop(togliere, axis = 1, inplace = True)
seriea = seriea.rename(columns = rename)
seriea
```

```python
seriea.to_csv(f'{cart}\\Serie-A.csv')
```

```python
fatto = ['Serie-A.csv']
```

```python
for file in os.listdir(cart):
    if ('keeper' not in file) and ('.csv' in file) and (file not in fatto):
        print(file)
        temp = pd.read_csv(f'{cart}\\{file}')
        temp.drop(togliere, axis = 1, inplace = True)
        temp = temp.rename(columns = rename)
        temp.to_csv(f'{cart}\\{file}')
        fatto.append(file)
```
