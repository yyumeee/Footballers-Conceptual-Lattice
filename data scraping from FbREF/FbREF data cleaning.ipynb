```python
ind = ['player', 'nationality', 'position', 'team', 'age', 'games']
turn_per90 = ['npxg_xg_assist', 'progressive_carries', 'progressive_passes', 'progressive_passes_received',
              'shots_free_kicks', 'passes_completed', 'passes', 'passes_short', 'passes_medium',
              'passes_long', 'pass_xa', 'key_passes', 'passes_into_final_third', 'passes_into_penalty_area',
              'crosses_into_penalty_area', 'passes_live', 'passes_dead', 'passes_free_kicks',
              'through_balls', 'passes_switches', 'crosses', 'throw_ins', 'corner_kicks', 'corner_kicks_in',
              'corner_kicks_out', 'passes_offsides', 'passes_blocked', 'sca_passes_live', 'sca_passes_dead',
              'sca_take_ons', 'sca_shots', 'sca_fouled', 'sca_defense', 'tackles', 'tackles_won',
              'tackles_def_3rd', 'tackles_mid_3rd', 'tackles_att_3rd', 'challenge_tackles', 'challenges',
              'blocks', 'interceptions', 'clearances', 'touches', 'touches_def_pen_area', 'touched_def_3rd',
              'touches_mid_3rd', 'touches_att_3rd', 'touches_att_pen_area', 'touches_live_ball',
              'take_ons', 'carries', 'carries_distance', 'passes_total_distance', 'passes_progressive_distance',
              'carries_progressive_distance', 'carries_into_final_third', 'carries_into_penalty_area',
              'miscontrols', 'dispossessed', 'passes_received', 'fouls', 'fouled', 'ball_recoveries',
              'aerials_won', 'aerials_lost']
min_avg = ['goals_per90', 'assists_per90', 'xg_per90', 'xg_assist_per90', 'npxg_per90',
           'shots_per90', 'shots_on_target_per90', 'sca_per90'
          ]
ratios = {'goals_per_shot': ['goals_per90', 'shots_per90'], 'goals_per_shot_on_target': ['goals_per90', 'shots_on_target_per90'],
          'npxg_per_shot': ['npxg_per90', 'shots_per90'], 
         }
pcts = {'shots_on_target_pct': ['shots_on_target_per90', 'shots_per90',], 'challenge_tackles_pct': ['challenge_tackles', 'challenges'],
       }
special = ['average_shot_distance', 'team', 'passes_pct', 'passes_pct_short', 'passes_pct_medium',
          'passes_pct_long', 'take_ons_won_pct', 'take_ons_tackled_pct', 'aerials_won_pct', 
          'player', 'position']
```

```python
playersdoubles = df_top5[['player', 'age', 'nationality']].value_counts()
playersdoubles = playersdoubles[playersdoubles > 1].index
df_top5.set_index(['player', 'age', 'nationality'], inplace = True)
dups = df_top5.loc[playersdoubles, :]
```

```python
dups['total_shot_distance'] = dups['average_shot_distance'] * dups['shots_per90'] * dups['minutes_90s']
```

```python
can_sum = list(set(df_top5.columns) - set(min_avg) - set(special) - set(ratios.keys()) - set(pcts.keys()))
can_sum.append('total_shot_distance')
len(df_top5.columns), (len(can_sum) + len(min_avg) + len(special) + len(ratios.keys()) + len(pcts.keys()))
```

```python
import warnings
warnings.simplefilter('ignore')
duplicates = pd.DataFrame(index = playersdoubles)
for col in can_sum:
    for pl in duplicates.index:
        v1, v2 = dups.loc[pl,col]
        duplicates.loc[pl, col] = v1 + v2
```

```python
def weighted_avg(val1, val2, w1, w2):
    tot1 = val1 * w1
    tot2 = val2 * w2
    pes_tot = w1+w2

    if pes_tot == 0:
        return 0
    else:
        return (tot1+tot2) / pes_tot
```

```python
for col in min_avg:
    for pl in duplicates.index:
        v1, v2 = dups.loc[pl, col]
        w1, w2 = dups.loc[pl, 'minutes_90s']
        duplicates.loc[pl, col] = round(weighted_avg(v1,v2,w1,w2), 1)
```

```python
for col, lista in ratios.items():
    duplicates[col] = round((duplicates[lista[0]] / duplicates[lista[1]]), 1)
```

```python
temp = pd.DataFrame(index = playersdoubles)
for col, lista in pcts.items():
    for pl in temp.index:
        v1 = duplicates.loc[pl, lista[0]]
        v2 = duplicates.loc[pl, lista[1]]
        if v2 == 0:
            temp.loc[pl,col] = 0
        else:
            temp.loc[pl, col] = round((100 * v1 / v2), 1)
duplicates = pd.concat([duplicates, temp], axis = 1)
```

```python
temp = pd.DataFrame(index = playersdoubles)

for col in ['passes_pct', 'passes_pct_short', 'passes_pct_medium','passes_pct_long']:
    for pl in temp.index:
        v1, v2 = dups.loc[pl, col]
        w1, w2 = dups.loc[pl, 'passes']
        temp.loc[pl, col] = round(weighted_avg(v1,v2,w1,w2), 1)


for col in ['take_ons_won_pct', 'take_ons_tackled_pct']:
    for pl in temp.index:
        v1, v2 = dups.loc[pl, col]
        w1, w2 = dups.loc[pl, 'take_ons']
        temp.loc[pl, col] = round(weighted_avg(v1,v2,w1,w2), 1)

duplicates = pd.concat([duplicates, temp], axis = 1)
```

```python
duplicates['average_shot_distance'] = round((duplicates['total_shot_distance'] / (duplicates['minutes_90s'] * duplicates['shots_per90'])),1)
```

```python
duplicates['aerials_won_pct'] = round((100*duplicates['aerials_won'] / (duplicates['aerials_won'] + duplicates['aerials_lost'])), 1)
```

```python
def position(v1, v2, w1, w2):
    if (v1 != v2) and (w2 > w1):
        return v2
    else:
        return v1

def team(v1,v2):
    if v1 != v2:
        return f'{v1}/{v2}'
    else:
        return v1
```

```python
temp = pd.DataFrame(index = playersdoubles)

for pl in temp.index:
    p1, p2 = dups.loc[pl, 'position']
    w1, w2 = dups.loc[pl, 'minutes_90s']
    temp.loc[pl, 'position'] = position(p1,p2,w1,w2)

    t1, t2 = dups.loc[pl, 'team']
    temp.loc[pl, 'team'] = team(t1, t2)

duplicates = pd.concat([duplicates, temp], axis = 1)
```

```python
duplicates.drop('total_shot_distance', axis = 1, inplace = True)
```
