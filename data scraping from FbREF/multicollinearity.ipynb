```python
import pandas as pd
import os

df = pd.read_csv('data\\top5_midfield.csv')

df.head()
```

```python
df = df[(df.games >= 20) & (df.minutes_90s >= 10)]
df.head()
```

```python
turn_per90 = ['npxg_xg_assist', 'progressive_carries', 'progressive_passes', 'progressive_passes_received',
              'shots_free_kicks', 'passes_completed', 'passes', 'passes_short', 'passes_medium',
              'passes_long', 'pass_xa', 'key_passes', 'passes_into_final_third', 'passes_into_penalty_area',
              'crosses_into_penalty_area', 'passes_live', 'passes_dead', 'passes_free_kicks',
              'through_balls', 'passes_switches', 'crosses', 'throw_ins', 'corner_kicks', 'corner_kicks_in',
              'corner_kicks_out', 'passes_offsides', 'passes_blocked', 'sca_passes_live', 'sca_passes_dead',
              'sca_take_ons', 'sca_shots', 'sca_fouled', 'sca_defense', 'tackles', 'tackles_won',
              'tackles_def_3rd', 'tackles_mid_3rd', 'tackles_att_3rd', 'challenge_tackles', 'challenges',
              'blocks', 'interceptions', 'clearances', 'touches', 'touches_def_pen_area', 'touches_def_3rd',
              'touches_mid_3rd', 'touches_att_3rd', 'touches_att_pen_area', 'touches_live_ball',
              'take_ons', 'carries', 'carries_distance', 'passes_total_distance', 'passes_progressive_distance',
              'carries_progressive_distance', 'carries_into_final_third', 'carries_into_penalty_area',
              'miscontrols', 'dispossessed', 'passes_received', 'fouls', 'fouled', 'ball_recoveries',
              'aerials_won', 'aerials_lost']
turn_per_match = ['goals', 'assists', 'cards_yellow', 'cards_red', 'errors', 'offsides', 'pens_won',
                  'pens_conceded', 'own_goals']

rename = {}

for col in turn_per90:
    rename[col] = f'{col}_per90'
    df[col] = df[col] / df['minutes_90s']
for col in turn_per_match:
    rename[col] = f'{col}_perGame'
    df[col] = df[col] / df['games']

df = df.rename(columns = rename)
```

```python
df['top5'] = True
ind = ['player', 'nationality', 'age', 'position', 'team', 'minutes_90s', 'top5']
df.set_index(ind, inplace = True)
df
```

```python
df2 = pd.read_csv('data\\other_midfield.csv')

df2.head()
```

```python
df2 = df2[(df2.games >= 20) & (df2.minutes_90s >= 10)]
df2.head()
```

```python
for col in turn_per90:
    df2[col] = df2[col] / df2['minutes_90s']
for col in turn_per_match:
    df2[col] = df2[col] / df2['games']

df2 = df2.rename(columns = rename)
```

```python
df2['top5'] = False
ind = ['player', 'nationality', 'age', 'position', 'team', 'minutes_90s', 'top5']
df2.set_index(ind, inplace = True)
df2
```

```python
df = df.rename(columns = ritrattare2)
```

```python
df.drop(togliere, axis = 1, inplace = True)
df
```

```python
import numpy as np
df.replace([np.inf, -np.inf], 0, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_stand = df.groupby(level = 'top5').apply(lambda x: pd.DataFrame(scaler.fit_transform(x), index = x.index, columns = x.columns))
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif_data = pd.DataFrame(index = df_stand.columns)
vif_data["VIF"] = [variance_inflation_factor(df_stand.values, i) for i in range(df_stand.shape[1])]
vif_data.sort_values('VIF', ascending = False).head()
```

```python
df_stand.drop('passes_per90', axis = 1, inplace = True)
vif_data = pd.DataFrame(index = df_stand.columns)
vif_data["VIF"] = [variance_inflation_factor(df_stand.values, i) for i in range(df_stand.shape[1])]
vif_data.sort_values('VIF', ascending = False).head()
```

```python
df_stand.drop('touches_per90', axis = 1, inplace = True)
vif_data = pd.DataFrame(index = df_stand.columns)
vif_data["VIF"] = [variance_inflation_factor(df_stand.values, i) for i in range(df_stand.shape[1])]
vif_data.sort_values('VIF', ascending = False).head()
```

```python
thresh = vif_data.VIF.max()
i = 3
while thresh > 10:
    print(f'iteration {i}')
    var_drop = vif_data[vif_data.VIF == vif_data.VIF.max()].index[0]
    
    i += 1
    print(f'{var_drop} da rimuovere')
    df_stand.drop(var_drop, axis = 1, inplace = True)

    vif_data = pd.DataFrame(index = df_stand.columns)
    vif_data["VIF"] = [variance_inflation_factor(df_stand.values, i) for i in range(df_stand.shape[1])]
    print(vif_data.sort_values('VIF', ascending = False).head())
    thresh = vif_data.VIF.max()
```

```python
def check_multicoll(df, threshold):
    i = 1
    vif_data = pd.DataFrame(index = df.columns)
    vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    thresh = vif_data.VIF.max()
    print(vif_data.sort_values('VIF', ascending = False).head(3))
    while thresh > threshold:
        print(f'iteration {i}')
        var_drop = vif_data[vif_data.VIF == vif_data.VIF.max()].index[0]
        
        i += 1
        print(f'{var_drop} da rimuovere')
        df.drop(var_drop, axis = 1, inplace = True)
    
        vif_data = pd.DataFrame(index = df.columns)
        vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
        print(vif_data.sort_values('VIF', ascending = False).head(3))
        thresh = vif_data.VIF.max()
```
